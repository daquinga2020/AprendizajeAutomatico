```python
from sklearn.preprocessing import OneHotEncoder
from sklearn import tree
import graphviz
import matplotlib.pyplot as plt

data = {
    'Prevision': ['nublado', 'lluvioso', 'nublado', 'lluvioso', 'soleado', 'lluvioso', 'soleado', 'soleado', 'nublado', 'soleado'],
    'Temperatura': ['media', 'alta', 'baja', 'media', 'alta', 'baja', 'baja', 'alta', 'baja', 'media'],
    'Marea': ['baja', 'media', 'media', 'alta', 'alta', 'baja', 'alta', 'alta', 'media', 'baja'],
    'Viento': ['medio', 'debil', 'fuerte', 'fuerte', 'debil', 'medio', 'fuerte', 'medio', 'fuerte', 'debil'],
    'Pescar': ['si', 'si', 'no', 'si', 'no', 'si', 'no', 'no', 'no', 'si']
}

# Datos de entrada con características categóricas
caracteristicas = [
    ['nublado', 'media', 'baja', 'medio'],
    ['lluvioso', 'alta', 'media', 'debil'],
    ['nublado', 'baja', 'media', 'fuerte'],
    ['lluvioso', 'media', 'alta', 'fuerte'],
    ['soleado', 'alta', 'alta', 'debil'],
    ['lluvioso', 'baja', 'baja', 'medio'],
    ['soleado', 'baja', 'alta', 'fuerte'],
    ['soleado', 'alta', 'alta', 'medio'],
    ['nublado', 'baja', 'media', 'fuerte'],
    ['soleado', 'media', 'baja', 'debil']
]

# Crear un codificador one-hot para cada característica categórica
encoder = OneHotEncoder(sparse=False)
caracteristicas_codificadas = encoder.fit_transform(caracteristicas)

nombres_caracteristicas = encoder.get_feature_names_out(['Prevision', 'Temperatura', 'Marea', 'Viento'])

print(caracteristicas_codificadas)
print(caracteristicas_codificadas)
# Definición de la salida
pescar = ['si', 'si', 'no', 'si', 'no', 'si', 'no', 'no', 'no', 'si']

# Crear un clasificador de árbol de decisión
clf = tree.DecisionTreeClassifier()

# Entrenar el modelo con los datos codificados
clf = clf.fit(caracteristicas_codificadas, pescar)


# Visualizar el árbol de decisión directamente en Python
plt.figure(figsize=(12, 8))
tree.plot_tree(clf, filled=True, feature_names=encoder.get_feature_names_out(['Prevision', 'Temperatura', 'Marea', 'Viento']))
plt.show()


```
